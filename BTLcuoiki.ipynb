# C,Loại bỏ cột tương quan

```python
import pandas as pd
import numpy as np
def drop_correlated_columns(df, threshold):
    """
    Loại bỏ các cột có độ tương quan cao với các cột khác trong DataFrame dựa vào ngưỡng threshold.
    
    Parameters:
    - df: DataFrame đầu vào
    - threshold: Ngưỡng tương quan để quyết định xem cột có nên bị loại bỏ hay không
    
    Returns:
    - DataFrame mới sau khi loại bỏ các cột không cần thiết
    """
    # Tính toán ma trận tương quan
    corr_matrix = df.corr().abs()
    
    # Tạo một ma trận boolean với True ở vị trí cần loại bỏ
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    # Tìm các cột cần loại bỏ
    to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > threshold)]
    
    # Loại bỏ các cột không cần thiết
    df_filtered = df.drop(to_drop, axis=1)
    
    return df_filtered

# Loại bỏ các cột có độ tương quan > 0.7
df_filtered = drop_correlated_columns(df1, 0.7)

df_filtered 
```

# D, Xử Lí Ngoại Lệ = phương pháp IQR

```python
import pandas as pd

def remove_outliers_iqr(df, threshold=1.5):
    """
    Loại bỏ các điểm ngoại lệ từ DataFrame sử dụng phương pháp IQR (khoảng giữa phần tư thứ nhất và thứ ba).

    Parameters:
    - df: DataFrame đầu vào
    - threshold: Ngưỡng để xác định điểm ngoại lệ. Một giá trị phổ biến là 1.5.

    Returns:
    - DataFrame mới sau khi loại bỏ điểm ngoại lệ
    """
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    return df[(df >= lower_bound) & (df <= upper_bound)].dropna()


# Loại bỏ điểm ngoại lệ từ toàn bộ dữ liệu
filtered_df = remove_outliers_iqr(df_filtered )

```

# F, Chuẩn hoá dữ liệu

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Chuẩn hóa dữ liệu
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

# Áp dụng PCA với số lượng thành phần chính mong muốn
n_components = 21
pca = PCA(n_components=n_components)
pca.fit(scaled_data)

# Biến đổi dữ liệu ban đầu thành các thành phần chính
pca_data = pca.transform(scaled_data)

# Tạo DataFrame mới từ các thành phần chính
pca_df = pd.DataFrame(data=pca_data, columns=[f'PC{i}' for i in range(1, n_components + 1)])
pca_df 


```

# II, Ứng dụng mô hình học máy cho dữ liệu bài toán

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron

# Assume 'X' is your feature matrix and 'y' is your target variable

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(pca_df, Y, test_size=0.2, random_state=42)

# In ra kích thước của tập huấn luyện và tập kiểm tra
print("Kích thước tập huấn luyện:", X_train.shape, y_train.shape)
print("Kích thước tập kiểm tra:", X_test.shape, y_test.shape)
```

```python

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load dữ liệu mẫu

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
# Tạo một bộ phân loại SVM
svm_classifier = SVC(kernel='linear')  # Chọn kernel tùy thuộc vào bài toán và dữ liệu

# Huấn luyện bộ phân loại trên tập huấn luyện
svm_classifier.fit(X_train, y_train)

# Dự đoán nhãn cho tập kiểm tra
y_pred = svm_classifier.predict(X_test)

# Đánh giá độ chính xác của mô hình
accuracy = accuracy_score(y_test, y_pred)

print("Ti le du doan dung theo Perceptron:")
print('accuracy_score:', accuracy_score(y_test,y_pred))
print('precision_score:', precision_score(y_test,y_pred, average="micro"))
print('recall_score:', recall_score(y_test,y_pred, average="micro"))
print('f1_score:', f1_score(y_test,y_pred, average="micro"))
print("Độ chính xác của mô hình:", accuracy)
```

```python
# Biểu đồ thể hiện độ chính xác của mô hình qua các chỉ số dự đoán

import matplotlib.pyplot as plt
metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
metrics_values = [accuracy_score(y_test, y_pred),
                  precision_score(y_test, y_pred, average="micro"),
                  recall_score(y_test, y_pred, average="micro"),
                  f1_score(y_test, y_pred, average="micro")]

plt.bar(metrics_names, metrics_values, color=['blue', 'green', 'red', 'purple'])
plt.title('Perceptron Model Evaluation')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.show()
```
